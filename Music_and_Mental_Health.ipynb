```python
#Make a class model named Music Recommendation
class MusicRecommendation:
  #Creat variable dictionary music_genres to put in top three genres for each mental state
  #This will be used to store and retrieve music genre names (values) later on, stored in mental state keys
    def collect_ratings(self):
        music_genres = {
            'Anxiety': ['Folk', 'Rock', 'Kpop'],
            'Depression': ['Lofi', 'Hiphop', 'Folk'],
            'Insomnia': ['Lofi', 'Gospel', 'Metal'],
            'Panic': ['Lofi', 'R&B', 'Rap']
        }

        #Create empty dictionary to store user ratings
        ratings = {}
        #Creat loop over variable (key: mental state) for users to type in mental state
        for state in music_genres.keys():
            #Use input function to take user inputs
            #Use state placeholder to place in mental health states in order
            rating = int(input(f"\nRate your {state} from 1 to 5: "))
            #Store the rating in dictionary
            ratings[state] = rating
        #Return dictionary with user ratings
        return ratings

    #Create function to recommend music referring to dictionary created earlier for  user ratings
    def recommend_music(self, ratings):
        #Dictionary to store genre scores for later calcultion
        genre_scores = {}
        music_genres = {
            'Anxiety': ['Folk', 'Rock', 'Kpop'],
            'Depression': ['Lofi', 'Hiphop', 'Folk'],
            'Insomnia': ['Lofi', 'Gospel', 'Metal'],
            'Panic': ['Lofi', 'R&B', 'Rap']
        }

        #Create loop to iterate over the mental state and the ratings they have, whch users have typed in
        for state, rating in ratings.items():
            #Take a look at top three genres for each mental health state
            genres = music_genres[state]
            #Use i and genre variable to find the index to calculcate the overall score of each genres in mental health state
            for i, genre in enumerate(genres):
                #Make if statement to score the genres. As the for loop goes, each genre would be scored as its placements in mental state
                if genre not in genre_scores:
                    genre_scores[genre] = 0
                genre_scores[genre] += (3 - i) * rating

        #Rank genres based on scores in descending order
        #Retrieve key from genre_scores dictonary, returning total score of each music genre in descending order
        #Used lambda since I heard that it's good for sorting
        sorted_genres = sorted(genre_scores.keys(), key=lambda x: genre_scores[x], reverse=True)

        #Get the top 3 genres as recommendations from sorted list in order to get music genres with the highest scores
        top_3_recommendations = sorted_genres[:3]

        return top_3_recommendations

#Use class created earlier to make object function
recommendation_system = MusicRecommendation()

#Collect ratings from users
user_ratings = recommendation_system.collect_ratings()

#Get optimal top 3 music genre recommendations based on ratings
optimal_recommendations = recommendation_system.recommend_music(user_ratings)

#Display recommendations as a result
print("\nOptimal Music Recommendations:")
for i in range(len(optimal_recommendations)):
    print(f"{i + 1}. {optimal_recommendations[i]}")

while True:
    #Ask if want to retake
    again = input("\nWould you like to take again? (Enter Yes/No) ")

    if again.lower() == "no":
        print("\nThanks for participating!")
        break

    elif again.lower() == "yes":
        #Collect ratings from users again
        user_ratings = recommendation_system.collect_ratings()

        #Get optimal top 3 music genre recommendations based on new ratings
        optimal_recommendations = recommendation_system.recommend_music(user_ratings)

        #Display new recommendations
        print("\nOptimal Music Recommendations:")
        for i in range(len(optimal_recommendations)):
            print(f"{i + 1}. {optimal_recommendations[i]}")
    else:
        print("\nInvalid input. Please enter Yes or No.")
```
